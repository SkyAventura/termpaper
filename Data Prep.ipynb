```python
import codecs
import numpy as np
import pandas as pd
```

```python
i = 0
#
# First load book ratings into self.data
#
data = {}
f = codecs.open("BX-Book-Ratingsc.csv", 'r', 'utf8')
for line in f:
    i += 1
    fields = line.split(';')
    user = fields[0].strip('"')
    book = fields[1].strip('"')
    rating = int(fields[2].strip().strip('"'))
    if user in data:
        currentRatings = data[user]
    else:
        currentRatings = {}
    currentRatings[book] = rating
    data[user] = currentRatings
f.close()
```

```python
df=pd.DataFrame.from_dict(data, orient='index')
```
